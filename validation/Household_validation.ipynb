```python
# plot the data itself
categories = ['two', 'three', 'four', 'five', 'six', 'seven or more']
for cat in categories:
    fig, ax = plt.subplots()
    print
    x1 = hh_size['%s_x'%(cat)]
    y1 = hh_size['%s_y'%(cat)]

    plt.plot(y1,x1,'y.')
    #plt.plot(y,x2,'b.')
    #plt.plot(x3,y,'r.')
    plt.axis([0,150,0,150])
    plt.xlabel('Synthetic Population')
    plt.ylabel('OFM weighted ACS')
    plt.title('Household Size: %s'%(cat))
    plt.show()
    s = np.std(x1)
    m = np.mean(x1)
    print 'Std=',s,'; Mean=', m
```
